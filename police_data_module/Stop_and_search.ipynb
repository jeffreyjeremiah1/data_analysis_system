```python
# deleting columns that i dont need
df.drop(["operation_name", "object_of_search"], axis = 1, inplace = True)
```

```python
# Reassign issue Date.
df["datetime"] = pd.to_datetime(df["datetime"])
df.head()
```

```python
# seperated ethnicity to one word 'black', 'white', 'mixed', and 'other'
df['self_defined_ethnicity'] = df['self_defined_ethnicity'].str.replace('-', ' ')
df['self_defined_ethnicity'] = df['self_defined_ethnicity'].str.replace('/', ' ')
df['self_defined_ethnicity'] = df['self_defined_ethnicity'].str.replace(',', ' ')
df['self_defined_ethnicity'] = df['self_defined_ethnicity'].str.split().str[0]

df.self_defined_ethnicity
```

```python
# creating a new column called month
df["month"] = df["datetime"].dt.month
df.head()
```

```python
# # creates new columns for nested data in location
df = df.join(df.pop('location').apply(pd.Series))

df = df.join(df.pop('street').apply(pd.Series))
df.describe
```

```python
# drop row where gender is 'Other'
df = df.drop(df[df['gender'] == 'Other'].index)
df.gender
```

# Questions

## 1. What are the top 5 age range with the most arrests ?

```python
# Group the data by age range and count the number of arrests
age_arrests = df.groupby('age_range')['outcome'].agg(lambda x: x[x == 'Arrest'].count())

# Sort the data in descending order by the count of arrests
age_arrests = age_arrests.sort_values(ascending=False)

# Select the top 5 age ranges
top_5_age_ranges = age_arrests.head(5)

# Print the top 5 age ranges
print(top_5_age_ranges)
```

```python
#  Plots a pie Chart    
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(14, 8), dpi=200, subplot_kw=dict(aspect="equal"))

data = top_5_age_ranges.values
age_ranges = top_5_age_ranges.index

def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d} Arrests)".format(pct, absolute)

wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))
ax.legend(wedges, age_ranges,
          title="Age Ranges",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
ax.set_title(f"Top 5 Age Ranges with highest Arrests", color='red', fontsize=24)

plt.show()
```

## 2. What are the top locations where interactions takes place ?

```python
# Group the data by location and count the number of interactions
location_count = df.groupby('name').size().sort_values(ascending=False)
top_10_locations = location_count.head(10)

plt.style.use('fivethirtyeight')

# Set dimensions for graph
fig = plt.figure(figsize=(14, 8), dpi=200)

# Create a bar chart to show the location count
ax = fig.add_subplot(111)
top_10_locations.plot(kind='bar', color='lightblue', edgecolor='black')

# Add a title and labels to the chart
ax.set_title('Top 10 Location names of interactions ', fontsize=22, color='red')
ax.set_xlabel('Location ', color='red')
ax.set_ylabel('Interaction Count', color='red')


plt.show()


```

## 3. Are there any patterns or disparities in the outcomes based on self-defined ethnicity and officer-defined ethnicity ?

```python
# Group the data by ethnicity and outcome and count the number of interactions
ethnicity_count = df.groupby(['self_defined_ethnicity', 'officer_defined_ethnicity', 'outcome']).size().unstack()
ethnicity_count
```

```python
# Group the data by ethnicity and outcome and count the number of interactions
ethnicity_count = df.groupby(['self_defined_ethnicity', 'officer_defined_ethnicity', 'outcome']).size().unstack()

plt.style.use('fivethirtyeight')

# Create a stacked bar chart to show the ethnicity count by outcome
ethnicity_count.plot(kind='bar', figsize=(14, 8), stacked=False, color=['lightblue','crimson', 'lightgreen', 'darkorange', 'teal', 'indigo'], edgecolor='black')

# Add a title and labels to the chart
plt.title('Ethnicity Count by Outcome', color='red', fontsize=24)
plt.xlabel('Ethnicity (self_defined_ethnicity, officer_defined_ethnicity)', color='red')
plt.ylabel('Count', color='red')
plt.xticks(rotation=0)
# Add a legend to the chart
plt.legend(ethnicity_count, title='Outcome')
```
